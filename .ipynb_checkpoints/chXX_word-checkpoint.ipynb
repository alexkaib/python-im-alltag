# Teil XX: Word-Dateien verarbeiten

```python
from docx import Document
```

## Absätze auslesen

```python
doc = Document("warenuebersicht.docx")
```

```python
for p in doc.paragraphs:
    print(p.text)
```

## Tabellen auslesen

```python
inventory = doc.tables[0]

inventory_cols = {}
for col in inventory.columns:
    col_name = col.cells[0].text
    inventory_cols[col_name] = []
    
    for cell in col.cells[1:]:
        inventory_cols[col_name].append(cell.text)

print(inventory_cols)
```

Übung: Funktion, die Tabelle und Spaltennamen nimmt und Summe der Spaltenwerte zurückgibt

```python
def sum_col(table, col_name):
    for col in table.columns:
        if col.cells[0].text == col_name:
            total = 0
            for cell in col.cells[1:]:
                total += int(cell.text)
            break
            
    return total
```

```python
sum_col(doc.tables[0], "Auf Lager")
```

## Auf Kopf- und Fußzeilen zugreifen

```python
s = doc.sections[0]

print(s.footer.paragraphs[0].text)
```

## Text ändern

Überschreiben des Text-Attributs

```python
doc.paragraphs[0].text = "Warenübersicht, Mai 2025"
```

Hinzufügen neuer Absätze

```python
doc.paragraphs[1].insert_paragraph_before("Autor: Peter Parker")
```

Speichern, ggf. unter neuem Namen

```python
doc.save("warenuebersicht_mai.docx")
```

Übung: "vergangenen Monat" durch "Mai" und "kommenden Monat" durch "Juni" ersetzen.

```python
for p in doc.paragraphs:
    p.text = p.text.replace("vergangenen Monat", "Mai").replace("kommenden Monat", "Juni")

doc.save("warenuebersicht_mai.docx")
```

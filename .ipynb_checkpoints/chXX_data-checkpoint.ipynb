## 1. Daten aufbereiten

Im Folgenden werden wir uns mit einem Datensatz von [Gapminder](https://www.gapminder.org/data/) beschäftigen, der das Bruttoinlandsprodukt pro Kopf (kaufkraft- und inflationsbereinigt) für verschiedene Länder über die letzten 225 Jahre angibt. Die Daten stammen aus verschiedenen Quellen und sind teils extrapoliert; nähere Informationen dazu finden sich [hier](https://www.gapminder.org/data/documentation/gd001/) und [hier](https://www.gapminder.org/data/documentation/).

Wir gehen zuerst der Frage nach, **wie sich das BIP in Deutschland in den letzten 125 entwickelt hat**. Das wird uns dabei helfen, die Funktionen von `pandas` Schritt für Schritt kennenzulernen

### Rohdaten einlesen

Die Daten liegen uns als `.csv`-Datei vor, ein unkompliziertes Rohtextformat, das alle vernünftigen Statistikprogramme verarbeiten können. pandas stellt uns die `read_csv`-Funktion zur Verfügung, um die Datei in unser Programm einzulesen. Dabei entsteht ein **Data Frame** (oft als **df** abgekürzt), eine für pandas eigens definierte Datenstruktur, die als Tabelle mit Zeilen und Spalten verstanden werden kann.

````python
# Einlesen der Rohdaten als Data Frame, den wir 'gdp' nennen
gdp = pd.read_csv('datasets/gapminder_gdp_pcap.csv')
````

### Überblick verschaffen

Die `head` und `info` Methoden liefern uns einen groben Eindruck, wie die Daten strukturiert sind.

````python
# head() liefert die ersten 5 Zeilen des Data Frame
gdp.head()
````

````python
# info() liefert die Dimensionen und Datentypen der Reihen und Spalten
gdp.info()
````

### Zeilenindex definieren

Standardmäßig werden die Zeilen eines Data Frame einfach durchnummeriert. In unserem Datensatz wäre es aber sinnvoll, die verschiedenen Länder als Zeilennamen zu verwenden. Dafür benutzen wir die `set_index`-Methode mit dem Namen der Spalte, die wir als Index benutzen möchten.

**Wichtig**: Die meisten `pandas`-Funktionen erzeugen standardmäßig einen neuen Data Frame, statt einen bestehenden zu verändern. Deshalb ist meist eine Neuzuweisung der Variable notwendig.

````python
# Wir setzen die Spalte "name" als Zeilen-Index, um über den Ländernamen auf die Daten zugreifen zu können
gdp = gdp.set_index('name')
gdp.head()
````

### Unnötige Spalten ausblenden

Die Spalte "geo" mit den Ländercodes ist für unsere Zwecke unnötig und stört bei Berechnungen. Mit dem `drop`-Befehl können wir sie entfernen.

````python
# Wir entfernen die Spalte "geo", um einheitlichen float-Daten zu erhalten
gdp = gdp.drop(columns=['geo'])
gdp.head()
````

### Auf bestimmte Ausschnitte zugreifen

Aus den Daten interessiert uns aktuell nur eine Zeile (Deutschland) und ein Abschnitt der Spalten (1900 - 2025). Das `.loc`-Konstrukt ist eine kompakte Möglichkeit, um auf solche Ausschnitte eines Data Frame zuzugreifen. Wir schauen uns die Syntax etwas genauer an:
```python
pf.loc['start_index':'stop_index', 'start_column':'stop_column']
```
Auffällig sind zunächst die **eckigen Klammern** - sie zeigen, dass wir hier ähnlich wir bei Listen und Dictionaries auf einen Ausschnitt einer Datenstruktur zugreifen. In den Klammern folgt zunächst das gewünschte **Zeilenintervall** und dann nach einem Komma das **Spaltenintervall**. Wir nutzen dabei die **Namen** der Zeilen und Spalten, um den Anfang und das Ende der Intervalle zu benennen. Statt Intervallen können auch nur einzelne Zeilen und Spalten genannt werden.

Nebenbemerkung: Das analoge `iloc`-Konstrukt würde uns erlauben, die Zeilen und Spalten über ihren **Index** statt über den Namen anzusprechen. In unserem Datensatz sind aber die Namen der intuitivere Ansatz.

````python
# Wir beschränken unsere Analyse auf Deutschland in den Jahren 1900 - 2025
de = gdp.loc['Germany', '1900':'2025']
de.head()
````

Weil wir in der neuen `de`-Variable jetzt nur noch eine einzelne Zeile speichern, liefert uns pandas eine **Serie** statt einen Data Frame. Diese Datenstruktur ähnelt einer Liste, besitzt aber viele zusätzliche statistische Methoden.

### Diagramme erzeugen

Sind die Daten in der richtigen Form, ist die Erstellung von Diagrammen ganz einfach:

````python
# Wir lassen und ein Diagram über die Daten zeichnen
de.plot()
````

Falls wir uns einen Abschnitt aus den Daten genauer anschauen möchten, können wir wieder das `loc`-Konstrukt verwenden.

````python
# Wir schauen uns die Daten der letzten 25 Jahre genauer an
de.loc['2000':'2025'].plot()
````

### Eigene Berechnungen durchführen

Schließlich schauen wir uns noch beispielhaft an, wie Data Frames mit Pythons Kontrollstrukturen kombiniert werden können, um eigene Berechnungen durchzuführen.

````python
# Wir berechnen, wie viele Länder mehr bzw. weniger BIPPP als Deutschland hatten
gdp_de = gdp2017['Germany']
more_than = 0
less_than = 0

for country in gdp2017.index:
    if country == 'Germany':
        continue
    elif gdp2017[country] > gdp_de:
        more_than += 1
    else:
        less_than += 1

print("Anzahl mehr:", more_than)
print("Anzahl weniger:", less_than)
print("Prozent mehr:", round(100 * more_than / len(gdp2017), 2))
print("Prozent weniger:", round(100 * less_than / len(gdp2017), 2))
````

## 3: Korrelationen finden

````python
# als Wert zw. 0 und 100 normalisierte Anzahl an durchschnittlichen Bildungsjahren
edu = pd.read_csv('datasets/gapminder_owid_education_idx.csv')
edu.head()
````

````python
# Vorherige Transformationsschritte in einem
edu = edu.set_index('name').drop(columns=['geo']).loc[:, '2017']
````

````python
edu
````

````python
# Wir kombinieren die zwei Serien zu einem Dataframe
gdp_edu = pd.concat([gdp2017, edu], axis=1, keys=['gdp', 'edu_idx'])
````

````python
gdp_edu.dropna()
````

````python
gdp_edu.plot(kind="scatter", x='gdp', y='edu_idx')
````

````python
for country in gdp_edu.index:
    if gdp_edu.loc[country, 'gdp'] < 20000 and gdp_edu.loc[country, 'edu_idx'] > 75:
        print(country, gdp_edu.loc[country, 'gdp'], gdp_edu.loc[country, 'edu_idx'])
````

````python
gdp_edu.corr()
````

````python
# Durchschnittliche wöchentliche Arbeitsstunden
hours = pd.read_csv('datasets/gapminder_working_hours_per_week.csv')
hours.head()
````

````python
hours = hours.set_index('name').loc[:, '2017']
````

````python
gdp_edu['hours'] = hours
````

````python
gdp_edu.plot(kind='scatter', x='hours', y='gdp')
````

````python
gdp_edu.corr()
````

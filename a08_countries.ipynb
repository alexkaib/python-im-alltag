## Programm

```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Python Tutorial Bot'
}

res = requests.get('https://de.wikipedia.org/wiki/Liste_der_Staaten_der_Erde', headers=headers)

countries_soup = BeautifulSoup(res.content, 'html.parser')
```

```python
import time

countries_table = countries_soup.find('table')

for row in countries_table.find_all('tr'):
    country = row.find('td')
    if not country or not country.find('a') or country.text.strip() == "Erde":
        continue
        
    name = country.text.strip()
    url = country.find('a')["href"]
    base_url = 'https://de.wikipedia.org'

    german_language_countries = 0
    
    try:
        country_page = requests.get(base_url + url, headers=headers)
        country_soup = BeautifulSoup(country_page.content, 'html.parser')
    
        infobox = country_soup.find('table', class_='infobox')
    
        for row in infobox.find_all('tr'):
            if "Amtssprache" in row.find('td').text:
                languages = row.find_all('td')[1].text.strip()
    
                if "Deutsch" in languages:
                    print(name)
                    german_language_countries += 1
                
                break
            
    except:
        print(f"Zugriff auf Sprache von {name} nicht erfolgreich")

    time.sleep(1)

print(german_language_countries)
```

```python
# Wir importieren die pandas Bibliothek und kürzen sie mit pd ab
import pandas as pd
```

```python
# Wir lesen unsere Rohdaten ein (mehr Infos zur Datenquelle: https://www.gapminder.org/data/documentation/gd001/)...
gdp = pd.read_csv('/Projekte/python-im-alltag/datasets/gapminder_gdp_pcap.csv')
# ... und schauen uns ihre Form an, indem wir die ersten 5 Zeilen ausgeben
gdp.head()
```

```python
# Wir setzen die Spalte "name" als Zeilen-Index, um über den Ländernamen auf die Daten zugreifen zu können
gdp = gdp.set_index('name')
gdp.head()
```

```python
# Wir entfernen die Spalte "geo", um einheitlichen float-Daten zu erhalten
gdp = gdp.drop(columns=['geo'])
gdp.head()
```

```python
# Wir beschränken unsere Analyse auf Deutschland in den Jahren 1900 - 2025
de = gdp.loc['Germany', '1900':'2025']
de
```

```python
# Wir lassen und einen Graphen über die Daten zeichnen
de.plot()
```

## Teil 2: Perzentil DE weltweit

```python
gdp.head()
```

```python
# Wir schauen uns alle Länder im Jahr 2017 an
gdp2017 = gdp.loc[:, '2017']
gdp2017
```

```python
# Mit "describe()" erhalten wir typische statistische Messwerte
gdp2017.describe()
```

```python
# Mit dem "kind" Parameter lassen sich verschiedene Diagrammtypen erstellen, die ggf. besser zu den Daten passen
gdp2017.plot(kind='hist')
```

```python
# Wir berechnen, wie viele Länder mehr bzw. weniger BPIPP als Deutschland hatten
gdp_de = gdp2017['Germany']
more_than = 0
less_than = 0

for country in gdp2017.index:
    if country == 'Germany':
        continue
    elif gdp2017[country] > gdp_de:
        more_than += 1
    else:
        less_than += 1

print("Anzahl mehr:", more_than)
print("Anzahl weniger:", less_than)
print("Prozent mehr:", round(100 * more_than / len(gdp2017), 2))
print("Prozent weniger:", round(100 * less_than / len(gdp2017), 2))
```
